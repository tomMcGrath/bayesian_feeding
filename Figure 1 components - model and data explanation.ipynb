```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

# Figure 1a part 1

```python
n = 8
x = np.linspace(-np.pi, np.pi, n)
y = np.linspace(-np.pi, np.pi, n)
X, Y = np.meshgrid(x,y)
U = Y
V = -X

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(X, Y, np.zeros(n), U, V, np.zeros(n),
          normalize=True,
          arrow_length_ratio=0.3)

n = 32
x = np.linspace(-np.pi, np.pi, n)
y = np.linspace(-np.pi, np.pi, n)
X, Y = np.meshgrid(x,y)

Z = 5 + np.cos(X) + np.sin(Y)
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.zaxis.set_major_formatter(plt.NullFormatter())

plt.savefig('paper_figures/fig1a_pt1.svg')

plt.show()
```

# Figure 1a part 2

```python
n = 8
x = np.linspace(-np.pi, np.pi, n)
y = np.linspace(-np.pi, np.pi, n)
X, Y = np.meshgrid(x,y)
U = -0.3*X
V = -0.1*Y

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.quiver(X, Y, np.zeros(n), U, V, np.zeros(n),
          normalize=True,
          arrow_length_ratio=0.3)

n = 32
x = np.linspace(-np.pi, np.pi, n)
y = np.linspace(-np.pi, np.pi, n)
X, Y = np.meshgrid(x,y)

Z = 5*np.exp(-np.sqrt(X**2 + Y**2)) + 5
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm)

ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.yaxis.set_major_formatter(plt.NullFormatter())
ax.zaxis.set_major_formatter(plt.NullFormatter())

plt.savefig('paper_figures/fig1a_pt2.svg')

plt.show()
```
